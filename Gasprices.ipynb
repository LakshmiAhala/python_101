```python
# READING AND SPLITTING 'gasprices.txt'
with open('gasprices.txt') as gasprices_file_obj:
  
  for each_line in gasprices_file_obj:
    name_1 = each_line.split(':')
    #print(name_1)

    name_2 = name_1[0]
  #print(name_2)

    char = name_2.split('-')
    #print(char)

# MONTH
    month = int(char[0])
    print('Month: ', month)

# DAY
    day = int(char[1])
    print('Day: ', day)

# YEAR
    year = int(char[2])
    print('Year: ', year)
    
# GASPRICE
    gasprice = name_1[1]
    print('Gasprice: ', gasprice)
```
